```python
#Figure1
from sklearn import linear_model
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score

import numpy as np
import math

df=pd.read_csv("insurance.csv")
df=df[df['smoker']=='no']
X=df[['age','bmi','children']]
Y=df['charges']
regr=linear_model.LinearRegression()
regr.fit(X,Y)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X['age'],X['bmi'],Y)
plt.show()
predict=regr.predict(X)
residual=predict-Y
print(r2_score(Y, predict))

residual=residual/math.sqrt((mse(Y,predict)))

residual=pd.DataFrame(residual)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X=X[(residual['charges']<0.5) & (residual['charges']>-0.5)]
Y=Y[(residual['charges']<0.5) & (residual['charges']>-0.5)]
ax.scatter(X['age'],X['bmi'],Y)
plt.show()
regr.fit(X,Y)
predict=regr.predict(X)
residual=predict-Y
print(r2_score(Y, predict))
```

```python
#Data loading of the smoker category
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv("insurance.csv")
df2=df[df['smoker']=='yes']
X=df2.index
Yn=df2['charges'].values
Ynnn=df2['charges'].values
df3=pd.read_csv("insurance.csv")
df4=df3[df3['smoker']=='yes']
tem=np.zeros((274,3))
Ybmi=np.zeros((274,1))
for i in range(len(Yn)):
    if Yn[i] > 31745.265:
        Ynnn[i]=1
    else:
        Ynnn[i]=-1
for i in range(len(X)):
    if int(Ynnn[i])==1:
        ind=X[i]
        a=df4['age'][ind]
        b=df4['charges'][ind]
        c=df4['bmi'][ind]
        tem[i][0]=c
        tem[i][1]=b
        tem[i][2]=a
        plt.scatter(a,c,color='red')
    else:
        ind=X[i]
        a=df4['age'][ind]
        b=df4['charges'][ind]
        c=df4['bmi'][ind]
        tem[i][0]=c
        tem[i][1]=b
        tem[i][2]=a
        plt.scatter(a,c,color='green')
```

```python
#Split training set and testing set
testX=tem[0:20]
testY=Ynnn[0:20]
temp=tem[20:275,0:2]
temptest=tem[0:20,0:2]
print(testX.shape)
print(temptest.shape)
Ynn=Ynnn[20:275]
Ybmi=tem[20:275,2]
Ybmitest=tem[0:20,2]
print(Ybmi.shape)
```

```python
testX=pd.DataFrame(testX)
testX.to_csv('Testyes.csv')
```

```python
#Applying GMM
from sklearn import mixture
clf=mixture.GaussianMixture(n_components=2,covariance_type='full').fit(temp,Ynn)
predictY=clf.predict(temp)
print(predictY)
n=0
for i in range(len(temp)):
    if predictY[i]==1:
        n=n+1
m=len(temp)-n
print(m)
print(n)
agelarge=np.zeros((m,1)).astype(int)
chargelarge=np.zeros((m,1))
agesmall=np.zeros((n,1)).astype(int)
chargesmall=np.zeros((n,1))
n1=0
m1=0
for i in range(len(temp)):
    if predictY[i]==1:
        a=temp[i][0]
        b=temp[i][1]
        agesmall[n1]=a
        chargesmall[n1]=b
        n1=n1+1
        plt.scatter(a,b,color='green')
    else:
        a=temp[i][0]
        b=temp[i][1]
        agelarge[m1]=a
        chargelarge[m1]=b
        m1=m1+1
        plt.scatter(a,b,color='red')
Xinput=np.zeros((len(temp),2))
Xtestinput=np.zeros((len(temptest),2))
print(m1)
print(n1)

```

```python
#Appling SVM
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
for i in range(len(Xinput)):
    Xinput[i][0]=temp[i][0]
    Xinput[i][1]=Ybmi[i]
for i in range(len(Xtestinput)):
    Xtestinput[i][0]=temptest[i][0]
    Xtestinput[i][1]=Ybmitest[i]
clf=GridSearchCV(SVC(kernel="linear"), param_grid={'C': np.logspace(-3,3,10)}).fit(Xinput,predictY)
Xtestinput
```

```python
c=clf.best_params_['C']
clflin=SVC(kernel='linear',C=c).fit(Xinput,predictY)
newY=clflin.predict(Xinput)
newY
```
